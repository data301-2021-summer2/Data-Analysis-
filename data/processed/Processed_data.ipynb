```python
import pandas as pd

df = pd.read_csv('outbreaks.csv')
df.head()
```

```python
# Droping all the non existing values.
processed_df = df.dropna(axis=0)
print(processed_df.shape)
```

```python
# Replacing strings to numbers

replacement = {"January" : 1, "February":2,"March":3, "April" : 4, "May":5, "June":6,
               "July": 7, "August":8, "September":9,"October":10,"November":11,"December" :12,
               "Confirmed": 1,"Suspected": 0}

processed_df.replace(replacement)
```
